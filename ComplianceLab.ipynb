<h1>Botium Toys: Scope, goals, and risk assessment report</h1>

<h2>Project Summary</h2>

<p>This project is based on an activity included in the Google Cybersecurity Certificate. All names/scenarios are fictional.</p>

<h2>Scenario</h2>
<p>Botium Toys is a small U.S. business that develops and sells toys. The business has a single physical location, which serves as its main office, a storefront, and a product warehouse. However, Botium Toy’s online presence has grown, attracting customers in the U.S. and abroad. As a result, their information technology (IT) department is under increasing pressure to support their online market worldwide. 

<p>The manager of the IT department has decided that an internal IT audit needs to be conducted. They express concerns about not having a solidified plan of action to ensure business continuity and compliance, as the business grows. They believe an internal audit can help better secure the company’s infrastructure and help them identify and mitigate potential risks, threats, or vulnerabilities to critical assets. The manager is also interested in ensuring that they comply with regulations related to internally processing and accepting online payments and conducting business in the European Union (E.U.).</p>

<p>The IT manager starts by implementing the National Institute of Standards and Technology Cybersecurity Framework (NIST CSF), establishing an audit scope and goals, listing assets currently managed by the IT department, and completing a risk assessment. The goal of the audit is to provide an overview of the risks and/or fines that the company might experience due to the current state of its security posture.</p>

<p>This project's task is to review the IT manager’s scope, goals, and risk assessment report. Then, perform an internal audit by completing a controls and compliance checklist. </p>

<h2>Scope and goals of the audit</h2>

<p><b>Scope:</b> The scope is defined as the entire security program at Botium Toys. Therefore, all assets need to be assessed alongside internal processes and procedures related to the implementation of controls and compliance best practices.
</p>

<p><b>Goals:</b> Assess existing assets and complete the controls and compliance checklist to determine which controls and compliance best practices need to be implemented to improve the company’s security posture.</p>

<h2>Current assets</h2>

<p>Assets managed by the IT Department include:
<ul>
    <li>On-premises equipment for in-office business needs</li>
    <li>Employee equipment: end-user devices (desktops/laptops, smartphones), remote workstations, headsets, cables, keyboards, mice, docking stations, surveillance cameras, etc.</li>
    <li>Storefront products available for retail sale on-site and online; stored in the company’s adjoining warehouse</li>
    <li>Management of systems, software, and services: accounting, telecommunication, database, security, e-commerce, and inventory management</li>
    <li>Internet access</li>
    <li>Internal network</li>
    <li>Data retention and storage</li>
    <li>Legacy system maintenance: end-of-life systems that require human monitoring</li>
</ul>
</p>

<h2>Risk assessment</h2>
<h3>Risk description</h3>

<p>Currently, there is inadequate management of assets. Additionally, the company does not have all the proper controls in place and may not be fully compliant with both U.S. and international regulations and standards.</p>

<h3>Control best practices</h3>
<p>The first of the five functions of the NIST CSF is <q>Identify</q>. The company will need to identify assets so it can properly manage them. Additionally, the company will need to classify existing assets and determine the impact of loss these might have on business continuity.</p>

<h3>Risk score</h3>
<p>Due to the lack of controls and adherence to compliance best practices, the risk score is fairly high.</p>
